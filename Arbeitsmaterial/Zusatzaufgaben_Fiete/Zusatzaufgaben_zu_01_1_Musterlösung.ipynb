**9. <br>
Bestimmen Sie mithilfe einer einfachen Operation wieviele vollständige Fußballmanschaften à 11 Spieler sich aus 12.345 Fußballspielern bilden lassen. Wieviele Spieler bleiben traurig zurück? <br>
Geben Sie die Ergebnisse mit `print()` aus.**

```python
n_Spieler = 12345
n_Spieler_pro_Mannschaft = 11

# Antwort:

# Die Aufgabe lässt sich mit Hilfe des Integerdivision-Operators "//" lösen.

n_Mannschaften = n_Spieler // n_Spieler_pro_Mannschaft
n_restliche_Spieler = n_Spieler%n_Mannschaften
print('Es lassen sich', n_Mannschaften, 'Mannschaften bilden.')
print('Es bleiben',  n_restliche_Spieler, 'Spieler traurig zurück. :/')
```
