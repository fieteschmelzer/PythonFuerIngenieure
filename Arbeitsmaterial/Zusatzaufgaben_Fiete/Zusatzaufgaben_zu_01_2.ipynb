**8. <br>
Da sowohl die Keys als auch die Values vom Typ *Integer* sind, lässt  sich die Bestimmung des Zahlworts rekursiv schachteln. D.h. ein Key führt zu einem Value, der wiederum als Key verwendet werden kann, der zu einem weiteren Value führt usw. (`6` führt zu `5` führt zu `4`). <br>
Bemerkenswerterweise führen alle unsere Zahlwörter nach einigen rekursiven Aufrufen zum Wert `4`, der wiederum zu sich selbst führt.**
* Führen Sie diesen geschachtelten Aufruf für den Wert `7` solange aus, bis der Value `4` beträgt und geben Sie das Ergebnis mit `print()` aus!<br>
* Tun Sie das Selbe auch für die Keys `11`, `13`, `20` und `21`!

```python
# Beispiel:
print(zahlwortlänge[zahlwortlänge[6]]) # = print(zahlwortlänge[5])
j = [7, 11, 13, 20, 21]

for k in j:
    seen = set()
    while k not in seen:
        seen.add(k)
        k = zahlwortlänge[k]
    print(k)  # first repeated value = start of loop


# Antwort:

```
